```python
import pandas as pd
```

```python
df = pd.read_csv('customer_shopping_behavior.csv')
df.info()
```

```python
df.describe(include='all')
```

```python
df.isnull().sum()
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
#convert column name to snake case
#used later also to check columns whenever needed
df.columns = df.columns.str.lower() #
df.columns = df.columns.str.replace(' ', '_')
df.columns = df.columns.str.replace('(', '')
df.columns = df.columns.str.replace(')', '')
```

```python
df.columns
```

```python
age_label = ['Young Adult', 'Adult', 'Mid Age', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = age_label)
```

```python
df[['age', 'age_group']].head()
```

```python
# Column purchase frequency days

frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping).astype('Int64')
```

```python
df[['purchase_frequency_days', 'frequency_of_purchases']].head(10)
```

```python
df[['discount_applied', 'promo_code_used']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# as it returned true, means both columns have same values
# So eliminate one
df = df.drop('promo_code_used', axis=1)
```

```python
# 1. Define your connection parameters
server = 'localhost'
database = 'customer_behaviour'
table_name = 'customer_purchases'

# 2. Build the connection string
params = urllib.parse.quote_plus(
    f'DRIVER={{ODBC Driver 17 for SQL Server}};'
    f'SERVER={server};'
    f'DATABASE={database};'
    f'Trusted_Connection=yes;'
)

# 3. Create the SQLAlchemy engine
engine = sa.create_engine(f"mssql+pyodbc:///?odbc_connect={params}")

# 4. Upload the DataFrame
try:
    df.to_sql(table_name, con=engine, index=False, if_exists='replace')
    print(f"Successfully imported {len(df)} rows into table '{table_name}'.")
except Exception as e:
    print("Error occurred: ", e)
```
